```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('bmi.csv')
df=df.iloc[:100,:2]
df
```

```python
model = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='complete')
model.fit(df)
plt.scatter(df.iloc[:, 0], df.iloc[:, 1], c=model.labels_)
plt.xlabel('height')
plt.ylabel('weight')
plt.show()
```

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
clustering = linkage(df, method='complete') 
plt.figure(figsize=(20,20)) 
dendrogram(clustering)# 덴드로그램으로 표현 
plt.show()
```

```python
data=pd.read_csv('bmi.csv')
```

```python
data=data.iloc[:100,:]
l_data = data
data
```

```python
le=LabelEncoder()
le.fit(data["label"])
data['label']=le.transform(data["label"])
data['label']
```

```python
le.classes_
```

```python
target=data['label']
data=data.drop(['label'], axis=1)
data.shape
```

```python
import seaborn as sns

ax=sns.scatterplot(x='height',y='weight', hue=le.inverse_transform(target), style=le.inverse_transform(target), data=data)
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

cluster_complete=linkage(data, method='complete', metric='euclidean')
```

```python
plt.figure(figsize=(20,20))
dn=dendrogram(cluster_complete)
```

```python
from scipy.cluster.hierarchy import fcluster
predict=fcluster(cluster_complete, 55, criterion='distance')
predict
```

```python
from sklearn.cluster import AgglomerativeClustering 
cluster= AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='complete')
pred=cluster.fit_predict(data)
```
