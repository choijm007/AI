```python
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus']=False # 마이너스가 표시가 안될때 이렇게 해주면 된다.

X = np.linspace(-np.pi, np.pi, 256) 
C = np.cos(X)
plt.title("LaTeX, 문자열로 tick label 정의")
plt.plot(X, C)
# 틱 라벨 문자열에는 $$ 사이에 LaTeX 수학 문자식을 리스트로 넣어
# 지정 핛 수 잇다.
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
[r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$']) 
plt.yticks([-1, 0, 1], ["Low", "Zero", "High"])
plt.grid(True) # 격자 표시.기본값 : False plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus']=False 

f1=plt.figure(figsize=(10,2))
plt.title('figure size: (10,2)')
plt.plot(np.random.randn(100))
f2=plt.gcf()
print(f1,id(f1))
print(f2,id(f2))
plt.show()
```

```python
x1=np.linspace(0.,5.,50)
x2=np.linspace(0.,2.,50)
y1=np.cos(2*np.pi*x1)*np.exp(-x1)
y2=np.cos(2*np.pi*x2)
plt.suptitle("MAIN")

f1=plt.figure(figsize=(10,8))

plt.subplot(2,1,1)
plt.plot(x1,y1,'yo--')
plt.title('A Tale of 2 subplot')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2) # plt.subplot(212)
plt.plot(x2, y2, 'r.-')
plt.title('B Tale of 2 subplot')
plt.xlabel('time (s)') 
plt.ylabel('Undamped')
plt.tight_layout() # plot 갂의 갂격을 자동 조정해준다. plt.show()
```

```python
fig, ax0 = plt.subplots() # 복수개의 subplot생성
ax1 = ax0.twinx() # twinx():x축을 공유하는 새로운 Axes객체 생성 
ax0.set_title("2개의 y축 핚 figure에서 사용하기")
ax0.plot([10, 5, 2, 9, 7], 'r-', label="y0")
ax0.set_ylabel("y0")
ax0.set_xlabel("공유되는 x축")
ax0.grid(False)
ax1.plot([100, 200, 220, 180, 120], 'g:', label="y1") 
ax1.set_ylabel("y1")
ax1.grid(False)
plt.show()
```

```python
irum = ['hong', 'kim', 'lee']
jumsu = [90, 70, 80]
n_data = len(irum) # 젂체 데이터 수는 4
index = np.arange(n_data) # 배열 생성 (0, 1, 2, 3) 
c=['r', 'g', 'b']
plt.bar(index, jumsu, color = c, tick_label = irum)
plt.show()
```

```python
y1 = [70, 90, 80]
y2 = [60, 70, 90]
x = np.arange(len(y1))
err1 = np.abs(np.random.randn(3)) + 5 # 임의로 지정 
err2 = np.abs(np.random.randn(3)) + 5 # 임의로 지정 
xlabel = ['국어', '영어', '수학']
plt.title("Bar Chart")
plt.bar(x, y1, yerr=err1, ecolor='yellow', width=0.3, color='red', align='center',label="점수")
plt.bar(x+0.4, y2, yerr=err2, width=0.3, color='blue', align='center', label="반평균")
plt.xticks(x+0.4/2, xlabel)
plt.ylim(0, 100) 
plt.xlabel("과목")
plt.ylabel("점수")
plt.legend()
plt.show()
```

```python
y1 = [70, 90, 80] 
y2 = [60, 70, 90]
x = np.arange(len(y1))
err1 = np.abs(np.random.randn(3)) + 5 # 임의로 지정
err2 = np.abs(np.random.randn(3)) + 5 # 임의로 지정
xlabel = ['국어', '영어', '수학']
plt.title("Barh Chart")
plt.barh(x, y1, height=0.3, xerr=err1, ecolor='yellow', color='red',
align='center', label="점수")
plt.barh(x+0.3, y2, height=0.3, xerr=err2, color='blue', align='center',
label="반평균") 
plt.yticks(x+0.3/2, xlabel)
plt.xlim(0, 100)
plt.xlabel("점수") 
plt.ylabel("과목") 
plt.legend() 
plt.show()
```
