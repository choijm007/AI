```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus']=False 

f1=plt.figure(figsize=(10,2))
plt.title('figure size: (10,2)')
plt.plot(np.random.randn(100))
f2=plt.gcf()
print(f1,id(f1))
print(f2,id(f2))
plt.show()
```

```python
x1=np.linspace(0.,5.,50)
x2=np.linspace(0.,2.,50)
y1=np.cos(2*np.pi*x1)*np.exp(-x1)
y2=np.cos(2*np.pi*x2)
plt.suptitle("MAIN")

f1=plt.figure(figsize=(10,8))

plt.subplot(2,1,1)
plt.plot(x1,y1,'yo--')
plt.title('A Tale of 2 subplot')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2) # plt.subplot(212)
plt.plot(x2, y2, 'r.-')
plt.title('B Tale of 2 subplot')
plt.xlabel('time (s)') 
plt.ylabel('Undamped')
plt.tight_layout() # plot    . plt.show()
```

```python
fig, ax0 = plt.subplots() #  subplot
ax1 = ax0.twinx() # twinx():x   Axes  
ax0.set_title("2 y  figure ")
ax0.plot([10, 5, 2, 9, 7], 'r-', label="y0")
ax0.set_ylabel("y0")
ax0.set_xlabel(" x")
ax0.grid(False)
ax1.plot([100, 200, 220, 180, 120], 'g:', label="y1") 
ax1.set_ylabel("y1")
ax1.grid(False)
plt.show()
```

```python
N = 30
np.random.seed(0)
x = np.random.rand(N)
y1 = np.random.rand(N)
y2 = np.random.rand(N)
y3 = np.pi * (15 * np.random.rand(N))**2 
plt.title("Bubble Chart")
plt.scatter(x, y1, c=y2, s=y3) # c:color, s:size 
plt.show()
```

```python
irum = ['hong', 'kim', 'lee']
jumsu = [90, 70, 80]
n_data = len(irum) #    4
index = np.arange(n_data) #   (0, 1, 2, 3) 
c=['r', 'g', 'b']
plt.bar(index, jumsu, color = c, tick_label = irum)
plt.show()
```

```python
y1 = [70, 90, 80]
y2 = [60, 70, 90]
x = np.arange(len(y1))
err1 = np.abs(np.random.randn(3)) + 5 #   
err2 = np.abs(np.random.randn(3)) + 5 #   
xlabel = ['', '', '']
plt.title("Bar Chart")
plt.bar(x, y1, yerr=err1, ecolor='yellow', width=0.3, color='red', align='center',label="")
plt.bar(x+0.4, y2, yerr=err2, width=0.3, color='blue', align='center', label="")
plt.xticks(x+0.4/2, xlabel)
plt.ylim(0, 100) 
plt.xlabel("")
plt.ylabel("")
plt.legend()
plt.show()
```

```python
y1 = [70, 90, 80] 
y2 = [60, 70, 90]
x = np.arange(len(y1))
err1 = np.abs(np.random.randn(3)) + 5 #  
err2 = np.abs(np.random.randn(3)) + 5 #  
xlabel = ['', '', '']
plt.title("Barh Chart")
plt.barh(x, y1, height=0.3, xerr=err1, ecolor='yellow', color='red',
align='center', label="")
plt.barh(x+0.3, y2, height=0.3, xerr=err2, color='blue', align='center',
label="") 
plt.yticks(x+0.3/2, xlabel)
plt.xlim(0, 100)
plt.xlabel("") 
plt.ylabel("") 
plt.legend() 
plt.show()
```
