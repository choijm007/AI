```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
x=iris.petal_length.values
plt.rcParams["font.family"] = 'AppleGothic'
sns.rugplot(x)
plt.title("iris 데이터 중 꽆잎의 길이에 대한")
plt.show()
```

```python
x = iris.petal_length.values
sns.kdeplot(x)
plt.rcParams["font.family"] = 'Apple Gothic' # 핚글처리 
plt.title("Iris 데이터 중, 꽃잎의 길이에 대핚 Kernel Density Plot")
plt.show()
```

```python
x = iris.petal_length.values
sns.distplot(x, kde=True, rug=True, vertical=True) 
plt.rcParams["font.family"] = 'AppleGothic' # 핚글처리 
plt.title("Iris 데이터 중, 꽃잎의 길이에 대핚 Dist Plot")
plt.show()
```

# countplot


```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.countplot(x='class',data=titanic)
plt.show()
```

```python
sns.countplot(x='class', hue='who',data=titanic)
plt.title('타이타닉호의 각 클래스별, 승객 수')
plt.show()
```

```python
sns.countplot(y='class', hue='who',data=titanic)
plt.title('타이타닉호의 각 클래스별, 승객 수')
plt.show()
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind='kde', color='green')
plt.rcParams["font.family"] = 'Malgun Gothic' # 핚글처리 
plt.suptitle("꽃받침의 길이와 넓이의 Joint Plot", y=1.02)
plt.show()
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind='hex', color='green')
plt.rcParams["font.family"] = 'Malgun Gothic' # 핚글처리 
plt.suptitle("꽃받침의 길이와 넓이의 Joint Plot", y=1.02)
plt.show()
```

```python
sns.jointplot(x="sepal_length", y="sepal_width", data=iris, kind='scatter', color='green')
plt.rcParams["font.family"] = 'Malgun Gothic' # 핚글처리 
plt.suptitle("꽃받침의 길이와 넓이의 Joint Plot", y=1.02)
plt.show()
```

```python
tips= sns.load_dataset('tips')
sns.barplot(x="day", y="total_bill", data=tips)
plt.title("요일 별, 젂체 팁")
plt.show()
```

```python
tips= sns.load_dataset('tips')
sns.barplot(x="day", y="total_bill", hue='sex',data=tips)
plt.title("요일 별, 젂체 팁")
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("요일 별 젂체 팁의 Box Plot") 
plt.show()
```

```python
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title("요일 별 젂체 팁의 Violin Plot")
plt.show()
```

```python
sns.stripplot(x="day", y="total_bill", data=tips, jitter=True) 
plt.title("요일 별 젂체 팁의 Strip Plot")
plt.show()
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
plt.title("요일 별 젂체 팁의 Swarm Plot") 
plt.show()
```

```python
sns.barplot(x="day", y="total_bill", hue="sex", data=tips)
plt.title("요일 별, 성별 젂체 팁의 Histogram") 
plt.show()
```
